### Para avaliarmos o comportamento da variável RENDA vamos construir uma tabela de frequências considerando as seguintes classes em salários mínimos (SM)
#### <font color='blue'>.</font>

Classes de renda:

<b>A</b> ► Acima de 25 SM

<b>B</b> ► De 15 a 25 SM

<b>C</b> ► De 5 a 15 SM

<b>D</b> ► De 2 a 5 SM

<b>E</b> ► Até 2 SM


Na construção das classes de renda, considerei que o salário mínimo na época da pesquisa era de <b>R$ 788,00</b>.

### 1º Definindo os intevalos das classes em reais (R$)

```python
sm = 788
classes = [0,2*sm, 5*sm, 15*sm, 25*sm, dados["Renda"].max()]
```

### 2º Definir os labels das classes

```python
labels = ["Classe E", "Classe D", "Classe C", "Classe B", "Classe A"]
```

### 3º Construindo a coluna de frequências

```python
"""
Aqui, a desigualdade de renda já pode ser vista no conjunto analisando,
com muita gente nas classes mais inferiores, e pouca gente nas superiores.
"""

frequencias = pd.value_counts(
    pd.cut( x = dados["Renda"],
           bins = classes,
           labels = labels,
           include_lowest = True))


frequencias = frequencias.sort_index(ascending = False) # Colocando a coluna em ordem alfabética

frequencias
```

### 4º Construindo a coluna de percentuais

```python
"""
Menos de 1% dos entrevistados estão na classe A!
"""

 percentual = pd.value_counts(
    pd.cut( x = dados["Renda"],
           bins = classes,
           labels = labels,
           include_lowest = True),
           normalize = True)

percentual = percentual.sort_index(ascending = False) # Colocando a coluna em ordem alfabética

percentual = percentual*100

percentual
```

### 5º Juntando as colunas de frequência e percentuais e ordenando as linhas de acordo com os labels das classes

```python
tabela = pd.DataFrame(
    {'Frequência': frequencias, 'Porcentagem (%)': percentual}
)
tabela.rename_axis(index = "Classes Econômicas", inplace = True)
tabela
```

### Gráfico para vizualizar a distribuição da população nas Classes de Renda

```python
# UTILIZANDO A BIBLIOTECA SEABORN E O MÉTODO CATPLOT PARA A CONSTRUÇÃO DE UM GRÁFICO DE DISTRIBUIÇÃO DE FREQUÊNCIAS:
# Notamos de maneira acentuada a desigualdade social entre os entrevistados...

freq = pd.cut( x = dados["Renda"],
           bins = classes,
           labels = labels,
           include_lowest = True)
freq = freq.to_frame()

freq.reset_index(inplace = True)
sns.catplot(x = "Renda", kind = "count", data= freq)
```

> ### Conclusões

Temos uma grande quantidade de pessoas nas classes econômicas mais baixas, enquanto poucas pessoas concentra-se nas partes mais altas da sociedade.
